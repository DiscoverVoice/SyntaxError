```python
def generate_text_with_llm(model_name, prompt):
    llm = vllm.LLM(model=model_name, 
                   max_model_len=2048,
                   gpu_memory_utilization=0.4)
    sampling_params = vllm.SamplingParams(
        temperature=0.1,
        top_p=0.95,
        max_tokens=1024,
        stop=["</s>", "Human:", "Assistant:"]
    )
    
    outputs = llm.generate([prompt], sampling_params)
    generated_code = outputs[0].outputs[0].text

    return generated_code
```

```python
system_prompt = """
You are a helpful assistant tasked with generating optimized and efficient C++ code for a strategic card game.
Follow all instructions precisely and ensure the code is efficient for large inputs and follows best practices.
"""
```

```python
prompt_parts = [
    """
    Please give me a code for merge sort in c++
    """
]
```

```python
response = generate_text_with_llm(model_name, system_prompt)
for i, part in enumerate(prompt_parts):
    input_text = response + f"\n\n### Part {i+1}: {part.strip()}"
    output = generate_text_with_llm(model_name, input_text)
    response += output + "\n"
    print(f"Response after part {i+1}:\n", response)
```
